```python
import pandas as pd
import matplotlib.pyplot as plt

# Compare results (example)
results = {
    'Model': ['User-Based CF', 'Item-Based CF'],
    'RMSE': [1.33, 1.44],
    'MAE': [0.94, 1.06]
}

df_results = pd.DataFrame(results)
print(df_results)

# Plot results
plt.figure(figsize=(8, 4))
plt.bar(df_results['Model'], df_results['RMSE'], color=['blue', 'orange'])
plt.title('RMSE Comparison: User-Based vs. Item-Based')
plt.ylabel('RMSE')
plt.show()
```
