```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt
import seaborn as sns

# ----------------------------------------------------
# Étape 1: Chargement et préparation des données
# ----------------------------------------------------
print("Étape 1: Préparation des données...")
ratings = pd.read_csv('../data/ratings_clean.csv')
movies = pd.read_csv('../data/movies_clean.csv')

# Division des données (70% entraînement, 30% test)
train_data, test_data = train_test_split(
    ratings,
    test_size=0.3,
    stratify=ratings['user_id'],
    random_state=42
)

# Matrice utilisateur-article
user_item_matrix = train_data.pivot_table(
    index='user_id',
    columns='movie_id',
    values='rating'
).fillna(0)

# ----------------------------------------------------
# Étape 2: Matrices de Similarité
# ----------------------------------------------------
print("\nÉtape 2: Calcul des matrices de similarité...")

# Similarité utilisateur-utilisateur (user-based)
print("Calcul de la similarité utilisateur-utilisateur...")
user_similarity = cosine_similarity(user_item_matrix)

# Similarité article-article (item-based)
print("Calcul de la similarité article-article...")
item_similarity = cosine_similarity(user_item_matrix.T)  # Transposée pour items

# Visualisation des matrices
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.heatmap(user_similarity[:10, :10], annot=True, cmap='YlGnBu')
plt.title('Similarité Utilisateur-Utilisateur (10 premiers)')

plt.subplot(1, 2, 2)
sns.heatmap(item_similarity[:10, :10], annot=True, cmap='YlOrRd')
plt.title('Similarité Article-Article (10 premiers)')

plt.tight_layout()
plt.savefig('similarity_matrices.png')
plt.show()

# ----------------------------------------------------
# Étape 3: Fonctions de Prédiction
# ----------------------------------------------------
def predict_user_based(user_id, movie_id, k=5):
    """Prédiction basée sur les utilisateurs similaires"""
    try:
        # Trouver les k utilisateurs les plus similaires
        similar_users = np.argsort(user_similarity[user_id-1])[::-1][1:k+1]
        
        # Obtenir les notes des utilisateurs similaires pour le film cible
        ratings = user_item_matrix.iloc[similar_users][movie_id]
        valid_ratings = ratings[ratings != 0]
        
        if len(valid_ratings) > 0:
            return valid_ratings.mean()
        else:
            # Retourner la moyenne de l'utilisateur si aucune note valide
            return user_item_matrix.loc[user_id].mean()
    except:
        return np.nan

def predict_item_based(user_id, movie_id, k=5):
    """Prédiction basée sur les articles similaires"""
    try:
        # Trouver les k articles les plus similaires
        movie_idx = user_item_matrix.columns.get_loc(movie_id)
        similar_items = np.argsort(item_similarity[movie_idx])[::-1][1:k+1]
        similar_movie_ids = user_item_matrix.columns[similar_items]
        
        # Obtenir les notes de l'utilisateur pour les articles similaires
        ratings = user_item_matrix.loc[user_id, similar_movie_ids]
        valid_ratings = ratings[ratings != 0]
        
        if len(valid_ratings) > 0:
            return valid_ratings.mean()
        else:
            # Retourner la moyenne du film si aucune note valide
            return user_item_matrix[movie_id].mean()
    except:
        return np.nan

# ----------------------------------------------------
# Étape 4: Évaluation des Modèles
# ----------------------------------------------------
print("\nÉtape 3: Évaluation des modèles...")

def evaluate_model(predict_func, model_name):
    actual = []
    predicted = []
    
    for _, row in test_data.iterrows():
        user_id = row['user_id']
        movie_id = row['movie_id']
        
        # Ne considérer que les paires existantes dans la matrice
        if user_id in user_item_matrix.index and movie_id in user_item_matrix.columns:
            pred = predict_func(user_id, movie_id)
            if not np.isnan(pred):
                actual.append(row['rating'])
                predicted.append(pred)
    
    rmse = np.sqrt(mean_squared_error(actual, predicted))
    mae = mean_absolute_error(actual, predicted)
    
    print(f"\nRésultats pour {model_name}:")
    print(f"RMSE: {rmse:.4f}")
    print(f"MAE: {mae:.4f}")
    print(f"Échantillons évalués: {len(actual)}/{len(test_data)}")
    
    return rmse, mae

# Évaluation des deux modèles
user_rmse, user_mae = evaluate_model(predict_user_based, "Filtrage Collaboratif (User-Based)")
item_rmse, item_mae = evaluate_model(predict_item_based, "Filtrage Collaboratif (Item-Based)")

# ----------------------------------------------------
# Étape 5: Exemple de Recommandation
# ----------------------------------------------------
def generate_recommendations(user_id, model_type='user', top_n=5):
    """Générer des recommandations pour un utilisateur"""
    # Films non notés par l'utilisateur
    user_ratings = user_item_matrix.loc[user_id]
    unrated_movies = user_ratings[user_ratings == 0].index
    
    predictions = []
    for movie_id in unrated_movies:
        if model_type == 'user':
            pred = predict_user_based(user_id, movie_id)
        else:
            pred = predict_item_based(user_id, movie_id)
        predictions.append((movie_id, pred))
    
    # Trier par note prédite décroissante
    predictions.sort(key=lambda x: x[1], reverse=True)
    
    # Retourner les top_n recommandations
    top_recommendations = []
    for movie_id, pred_rating in predictions[:top_n]:
        movie_title = movies[movies['movie_id'] == movie_id]['title'].values[0]
        top_recommendations.append((movie_title, pred_rating))
    
    return top_recommendations

# Exemple: Recommandations pour l'utilisateur 1
print("\nExemple de recommandations pour l'utilisateur 1:")
user_id = 1
user_recs = generate_recommendations(user_id, model_type='user')
for i, (title, rating) in enumerate(user_recs, 1):
    print(f"{i}. {title} (Note prédite: {rating:.2f})")

# ----------------------------------------------------
# Étape 6: Sauvegarde des Résultats
# ----------------------------------------------------
results = pd.DataFrame({
    'Modèle': ['User-Based', 'Item-Based'],
    'RMSE': [user_rmse, item_rmse],
    'MAE': [user_mae, item_mae]
})

results.to_csv('resultats_modeles.csv', index=False)
print("\nRésultats sauvegardés dans 'resultats_modeles.csv'")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import mean_squared_error, mean_absolute_error

# ----------------------------------------------------
# 1. Chargement et préparation des données
# ----------------------------------------------------
print("Chargement et préparation des données...")
ratings = pd.read_csv('../data/ratings_clean.csv')
movies = pd.read_csv('../data/movies_clean.csv')

# Vérifier et corriger le mapping ID film -> Titre
print("Vérification du mapping film ID -> Titre...")
movie_id_to_title = {}
missing_titles = 0

for _, row in movies.iterrows():
    movie_id_to_title[row['movie_id']] = row['title']

print(f"Total films dans le dictionnaire: {len(movie_id_to_title)}")
print(f"Exemple de mapping: 1 -> {movie_id_to_title.get(1, 'Non trouvé')}")

# Division des données (80% entraînement, 20% test)
train_data, test_data = train_test_split(
    ratings,
    test_size=0.2,
    stratify=ratings['user_id'],
    random_state=42
)

# Création de la matrice utilisateur-article
print("\nCréation de la matrice utilisateur-article...")
user_item_matrix = train_data.pivot_table(
    index='user_id',
    columns='movie_id',
    values='rating'
).fillna(0)

# Vérifier les films dans la matrice
matrix_movie_ids = set(user_item_matrix.columns)
known_movie_ids = set(movie_id_to_title.keys())
missing_in_matrix = matrix_movie_ids - known_movie_ids
print(f"Films dans la matrice: {len(matrix_movie_ids)}, Films connus: {len(known_movie_ids)}")
print(f"Films sans titre dans la matrice: {len(missing_in_matrix)}")

# ----------------------------------------------------
# 2. Calcul et visualisation des matrices de similarité
# ----------------------------------------------------
print("\nCalcul des matrices de similarité...")

# Matrice article-utilisateur (transposée)
item_user_matrix = user_item_matrix.T

# 2a. Similarité utilisateur-utilisateur
user_similarity = cosine_similarity(user_item_matrix)
print("Similarité utilisateur-utilisateur calculée")

# 2b. Similarité article-article
item_similarity = cosine_similarity(item_user_matrix)
print("Similarité article-article calculée")

# 2c. Visualisation des matrices
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
sns.heatmap(user_similarity[:10, :10], annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matrice de Similarité Utilisateur-Utilisateur (Top 10)')

plt.subplot(1, 2, 2)
sns.heatmap(item_similarity[:10, :10], annot=True, cmap='viridis', fmt='.2f')
plt.title('Matrice de Similarité Article-Article (Top 10)')

plt.tight_layout()
plt.savefig('similarity_matrices.png', dpi=300)
print("Visualisation des matrices sauvegardée dans 'similarity_matrices.png'")
plt.show()

# ----------------------------------------------------
# 3. Fonctions de prédiction
# ----------------------------------------------------
def predict_user_based(user_id, movie_id, k=10):
    """Prédiction basée sur les utilisateurs similaires"""
    try:
        # Vérifier si l'utilisateur existe
        if user_id not in user_item_matrix.index:
            return np.nan
        
        # Trouver les k utilisateurs les plus similaires
        user_idx = user_item_matrix.index.get_loc(user_id)
        similar_users = np.argsort(user_similarity[user_idx])[::-1][1:k+1]
        
        # Obtenir les notes des utilisateurs similaires
        similar_ratings = user_item_matrix.iloc[similar_users][movie_id]
        
        # Filtrer les notes valides (non nulles)
        valid_ratings = similar_ratings[similar_ratings != 0]
        
        if len(valid_ratings) > 0:
            return valid_ratings.mean()
        elif not np.isnan(user_avg := user_item_matrix.loc[user_id].mean()):
            return user_avg
        else:
            return train_data['rating'].mean()
    except KeyError:
        return np.nan

def predict_item_based(user_id, movie_id, k=10):
    """Prédiction basée sur les articles similaires"""
    try:
        # Vérifier si l'article existe
        if movie_id not in item_user_matrix.index:
            return np.nan
        
        # Trouver l'index de l'article dans la matrice
        movie_idx = item_user_matrix.index.get_loc(movie_id)
        
        # Trouver les k articles les plus similaires
        similar_items = np.argsort(item_similarity[movie_idx])[::-1][1:k+1]
        similar_movie_ids = item_user_matrix.index[similar_items]
        
        # Obtenir les notes de l'utilisateur pour ces articles
        user_ratings = user_item_matrix.loc[user_id, similar_movie_ids]
        
        # Filtrer les notes valides (non nulles)
        valid_ratings = user_ratings[user_ratings != 0]
        
        if len(valid_ratings) > 0:
            return valid_ratings.mean()
        elif not np.isnan(item_avg := item_user_matrix.loc[movie_id].mean()):
            return item_avg
        elif not np.isnan(user_avg := user_item_matrix.loc[user_id].mean()):
            return user_avg
        else:
            return train_data['rating'].mean()
    except KeyError:
        return np.nan

# ----------------------------------------------------
# 4. Évaluation des modèles
# ----------------------------------------------------
print("\nÉvaluation des modèles...")

def evaluate_model(predict_func, model_name, k=10):
    """Évalue un modèle de recommandation"""
    # Filtrer les données de test valides
    valid_test = test_data[
        (test_data['user_id'].isin(user_item_matrix.index)) & 
        (test_data['movie_id'].isin(user_item_matrix.columns))
    ]
    
    actual = []
    predicted = []
    skipped = 0
    
    for _, row in valid_test.iterrows():
        pred = predict_func(row['user_id'], row['movie_id'], k)
        if not np.isnan(pred):
            actual.append(row['rating'])
            predicted.append(pred)
        else:
            skipped += 1
    
    if len(actual) == 0:
        return np.nan, np.nan, 0, skipped
    
    rmse = np.sqrt(mean_squared_error(actual, predicted))
    mae = mean_absolute_error(actual, predicted)
    return rmse, mae, len(actual), skipped

# Évaluation des deux modèles
print("Évaluation du modèle User-Based...")
user_rmse, user_mae, user_count, user_skipped = evaluate_model(predict_user_based, "User-Based")
print("Évaluation du modèle Item-Based...")
item_rmse, item_mae, item_count, item_skipped = evaluate_model(predict_item_based, "Item-Based")

# Affichage des résultats
print("\nRÉSULTATS D'ÉVALUATION")
print("=" * 50)
print(f"Modèle User-Based:")
print(f"- Échantillons évalués: {user_count}/{len(test_data)}")
print(f"- Prédictions ignorées: {user_skipped}")
print(f"- RMSE: {user_rmse:.4f}")
print(f"- MAE: {user_mae:.4f}")
print("\n" + "-" * 50)
print(f"Modèle Item-Based:")
print(f"- Échantillons évalués: {item_count}/{len(test_data)}")
print(f"- Prédictions ignorées: {item_skipped}")
print(f"- RMSE: {item_rmse:.4f}")
print(f"- MAE: {item_mae:.4f}")
print("=" * 50)

# ----------------------------------------------------
# 5. Système de recommandation avec titres de films
# ----------------------------------------------------
def get_movie_title(movie_id):
    """Retourne le titre du film ou un message par défaut"""
    return movie_id_to_title.get(movie_id, f"Film inconnu (ID: {movie_id})")

def generate_recommendations(user_id, model_type='user', top_n=5):
    """Génère des recommandations personnalisées avec titres de films"""
    # Vérifier si l'utilisateur existe
    if user_id not in user_item_matrix.index:
        print(f"Utilisateur {user_id} non trouvé dans le système")
        return []
    
    # Articles non notés par l'utilisateur
    user_ratings = user_item_matrix.loc[user_id]
    unrated_items = user_ratings[user_ratings == 0].index
    
    print(f"\nGénération de recommandations pour l'utilisateur {user_id} ({model_type}-based)")
    print(f"Films non notés: {len(unrated_items)}")
    
    predictions = []
    for movie_id in unrated_items:
        if model_type == 'user':
            pred = predict_user_based(user_id, movie_id)
        else:
            pred = predict_item_based(user_id, movie_id)
        
        if not np.isnan(pred):
            movie_title = get_movie_title(movie_id)
            predictions.append((movie_id, movie_title, pred))
    
    # Trier par note prédite
    predictions.sort(key=lambda x: x[2], reverse=True)
    return predictions[:top_n]

# ----------------------------------------------------
# 6. Exemple de recommandations détaillées
# ----------------------------------------------------
def display_recommendations(user_id, model_type='user', num_rec=50):
    """Affiche les recommandations avec explication"""
    recommendations = generate_recommendations(user_id, model_type, num_rec)
    
    if not recommendations:
        print(f"Aucune recommandation disponible pour l'utilisateur {user_id}")
        return
    
    print(f"\nTop {num_rec} recommandations pour l'utilisateur {user_id} ({model_type}-based):")
    print("-" * 70)
    print(f"{'Position':<10}{'ID Film':<10}{'Titre':<50}{'Note Prédite':<12}")
    print("-" * 70)
    
    for i, (movie_id, title, pred_rating) in enumerate(recommendations, 1):
        print(f"{i:<10}{movie_id:<10}{title:<50}{pred_rating:<12.2f}")
    
    # Affichage des détails pour la première recommandation
    if recommendations:
        movie_id, title, pred_rating = recommendations[0]
        print("\nExplication de la recommandation principale:")
        print(f"Film: {title} (ID: {movie_id})")
        print(f"Note prédite: {pred_rating:.2f}")
        
        if model_type == 'user':
            # Trouver les utilisateurs similaires qui ont aimé ce film
            user_idx = user_item_matrix.index.get_loc(user_id)
            similar_users = np.argsort(user_similarity[user_idx])[::-1][1:6]
            
            # Obtenir les notes des utilisateurs similaires
            similar_ratings = []
            for sim_user_idx in similar_users:
                sim_user_id = user_item_matrix.index[sim_user_idx]
                rating = user_item_matrix.iloc[sim_user_idx][movie_id]
                if rating > 0:
                    similar_ratings.append((sim_user_id, rating))
            
            if similar_ratings:
                print("\nRecommandé car apprécié par des utilisateurs similaires:")
                for user_id, rating in similar_ratings:
                    print(f"- Utilisateur {user_id} a donné une note de {rating}")
            else:
                print("\nRecommandé basé sur vos préférences générales")
        else:
            # Trouver les films similaires que l'utilisateur a aimés
            try:
                movie_idx = item_user_matrix.index.get_loc(movie_id)
                similar_items = np.argsort(item_similarity[movie_idx])[::-1][1:6]
                similar_movie_ids = item_user_matrix.index[similar_items]
                
                user_ratings = []
                for sim_movie_id in similar_movie_ids:
                    rating = user_item_matrix.loc[user_id, sim_movie_id]
                    if rating > 0:
                        movie_title = get_movie_title(sim_movie_id)
                        user_ratings.append((sim_movie_id, movie_title, rating))
                
                if user_ratings:
                    print("\nRecommandé car similaire à des films que vous avez aimés:")
                    for movie_id, title, rating in user_ratings:
                        print(f"- {title} (Note: {rating})")
                else:
                    print("\nRecommandé basé sur la popularité du film")
            except Exception as e:
                print(f"\nRecommandé basé sur les caractéristiques du film: {str(e)}")

# ----------------------------------------------------
# 7. Exécution principale
# ----------------------------------------------------
if __name__ == "__main__":
    # Choisir un utilisateur existant dans la matrice
    sample_user = 1
    print(f"Utilisateur {sample_user} dans la matrice: {sample_user in user_item_matrix.index}")
    
    # Générer et afficher les recommandations
    display_recommendations(sample_user, 'user')
    display_recommendations(sample_user, 'item')
    
    # Sauvegarder les résultats
    results = pd.DataFrame({
        'Modèle': ['User-Based', 'Item-Based'],
        'RMSE': [user_rmse, item_rmse],
        'MAE': [user_mae, item_mae],
        'Échantillons': [user_count, item_count],
        'Ignorés': [user_skipped, item_skipped]
    })
    
    results.to_csv('resultats_recommandation.csv', index=False)
    print("\nRésultats sauvegardés dans 'resultats_recommandation.csv'")
```
